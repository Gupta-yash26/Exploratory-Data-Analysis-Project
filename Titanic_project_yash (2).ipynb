*** Q3. did the survival rate differ by gender / if so,how much.***

```python
dataset.head()
```

```python

survival_rate = (dataset.groupby('Sex')['Survived'].mean()*100).round(1)


sns.barplot(x='Sex', y='Survived', data=dataset, ci=None, palette='Set1')
plt.title("Survival rate by gender")
plt.show()

survived = dataset[dataset['Survived'] == 1]


gender_counts = survived['Sex'].value_counts()


palette = "Set1"
colors = sns.color_palette(palette)

labels = ['Female','Male']
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Gender of Survivors')
plt.show()
```
